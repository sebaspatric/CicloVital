```python
from graphviz import Digraph
from IPython.display import Image

dot = Digraph(comment='Mapa Conceptual - Teoría Psicosexual de Freud', format='png')
dot.attr(rankdir='TB', size='100')

# Nodo central
dot.node('A', 'Perspectiva psicoanalítica del desarrollo humano - Sigmund Freud', shape='box', style='filled', fillcolor='lightblue')
#dot.node('B', 'Sigmund Freud', shape='ellipse')
#dot.edge('A', 'B', label='propuesta por')

# Fundamentos
dot.node('C', 'Fundamentos', shape='box', style='filled')
dot.edge('A', 'C', label='se fundamenta en')
dot.node('C1', 'Inconsciente, impulsos sexuales, infancia', shape='note')
dot.node('C2', 'Zonas erógenas como base del desarrollo', shape='note')
dot.edge('C', 'C1')
dot.edge('C1', 'C2')

# Aportes
dot.node('E', 'Aportes', shape='box', style='filled')
dot.edge('A', 'E', label='aporta')
dot.node('E1', 'Estructura: ello, yo, superyó', shape='note')
dot.node('E2', 'Inconsciente como motor', shape='note')
dot.node('E3', 'Sexualidad y personalidad', shape='note')
dot.edge('E', 'E1')
dot.edge('E1', 'E2')
dot.edge('E2', 'E3')

# Críticas
dot.node('F', 'Críticas', shape='box', style='filled')
dot.edge('A', 'F', label='recibe críticas')
dot.node('F1', 'Enfoque excesivo en sexualidad', shape='note')
dot.node('F2', 'Falta de evidencia empírica', shape='note')
dot.node('F3', 'Visión androcéntrica criticada', shape='note')
dot.edge('F', 'F1')
dot.edge('F1', 'F2')
dot.edge('F2', 'F3')

# Etapas del desarrollo (agrupadas en subgráfico con borde rojo)
dot.node('D', 'Etapas del desarrollo psicosexual', shape='box', style='filled', fillcolor='lightgrey')
dot.edge('A', 'D', label='incluye')

etapas = [
    ('D1', 'Etapa Oral (0-1 año)', 'Exploración oral de objetos y sonidos.'),
    ('D2', 'Etapa Anal (1-3 años)', 'Control de esfínteres y rutinas.'),
    ('D3', 'Etapa Fálica (3-6 años)', 'Curiosidad por roles y género.'),
    ('D4', 'Etapa de Latencia (6-12 años)', 'Desarrollo social y académico.'),
    ('D5', 'Etapa Genital (12+ años)', 'Maduración y relaciones afectivas.')
]

prev = None
with dot.subgraph(name='cluster_etapas') as etapa_cluster:
    etapa_cluster.attr(label='Etapas del desarrollo psicosexual',
                       #style='filled',
                      style='dashed',
                       color='red',
                       #color='whitesmoke',
                       fillcolor='mistyrose',
                       fontcolor='',
                       fontsize='12')
    for node_id, label, ejemplo in etapas:
        ejemplo_id = 'ex_' + node_id
        etapa_cluster.node(node_id, label, shape='box', fillcolor='white')
        etapa_cluster.node(ejemplo_id, f'Ej: {ejemplo}', shape='note')
        etapa_cluster.edge(node_id, ejemplo_id)
        if prev:
            dot.edge(prev, node_id)
        else:
            dot.edge('D', node_id)
        prev = node_id

# Autores posteriores
dot.node('G', 'Autores posteriores', shape='box', style='filled')
dot.edge('A', 'G', label='continuada por')
dot.node('G1', 'Anna Freud:\nmecanismos de defensa', shape='note')
dot.node('G2', 'Melanie Klein:\nrelaciones objetales', shape='note')
dot.node('G3', 'John Bowlby:\nteoría del apego', shape='note')
dot.node('G4', 'Nancy Chodorow:\npsicoanálisis feminista', shape='note')
dot.node('G5', 'Bollas/Ogden:\ninconsciente relacional', shape='note')
dot.edge('G', 'G1')
dot.edge('G1', 'G2')
dot.edge('G2', 'G3')
dot.edge('G3', 'G4')
dot.edge('G4', 'G5')

# Renderizar y mostrar
output_path = dot.render('mapa_teoria_psicosexual_freud_mejorado', cleanup=False)
Image(filename=output_path)
```

```python
from graphviz import Digraph
from IPython.display import Image

dot = Digraph(comment='Mapa Conceptual - Teoría Psicosocial de Erikson', format='png')
dot.attr(rankdir='TB', size='100')

# Nodo principal
dot.node('A', 'Perspectiva psicosocial del desarrollo humano - Erik Erikson', shape='box', style='filled', fillcolor='lightblue')
#dot.node('B', 'Erik Erikson', shape='ellipse')
#dot.edge('A', 'B', label='propuesta por')

# Fundamentos
dot.node('C', 'Fundamentos', shape='box', style='filled')
dot.edge('A', 'C', label='se fundamenta en')
dot.node('C1', 'Desarrollo influido por lo social y cultural', shape='note')
dot.node('C2', 'Crisis psicosocial en cada etapa', shape='note')
dot.edge('C', 'C1')
dot.edge('C1', 'C2')

# Aportes
dot.node('E', 'Aportes', shape='box', style='filled')
dot.edge('A', 'E', label='aporta')
dot.node('E1', 'Desarrollo durante toda la vida', shape='note')
dot.node('E2', 'Identidad como construcción social', shape='note')
dot.node('E3', 'Importancia del entorno', shape='note')
dot.edge('E', 'E1')
dot.edge('E1', 'E2')
dot.edge('E2', 'E3')

# Críticas
dot.node('F', 'Críticas', shape='box', style='filled')
dot.edge('A', 'F', label='recibe críticas')
dot.node('F1', 'Etapas no siempre son lineales', shape='note')
dot.node('F2', 'Falta de evidencia empírica', shape='note')
dot.node('F3', 'Resolución variable según cultura', shape='note')
dot.edge('F', 'F1')
dot.edge('F1', 'F2')
dot.edge('F2', 'F3')

# Autores posteriores
dot.node('G', 'Autores posteriores', shape='box', style='filled')
dot.edge('A', 'G', label='continuada por')
dot.node('G1', 'James Marcia:\nEstados de identidad', shape='note')
dot.node('G2', 'Daniel Levinson:\nTransiciones adultas', shape='note')
dot.node('G3', 'Joan Erikson:\nNovena etapa: sabiduría', shape='note')
dot.node('G4', 'Robert Kegan:\nConciencia y desarrollo', shape='note')
dot.node('G5', 'Carol Hoare:\nIdentidad vocacional y cultura', shape='note')
dot.edge('G', 'G1')
dot.edge('G1', 'G2')
dot.edge('G2', 'G3')
dot.edge('G3', 'G4')
dot.edge('G4', 'G5')

# Etapas del desarrollo agrupadas
dot.node('D', 'Etapas del desarrollo', shape='box', style='filled', fillcolor='lightgrey')
dot.edge('A', 'D', label='incluye')

# === Subgráficos por grupo etario ===
etapas = [
    ('E1a', 'Confianza vs. Desconfianza\n(0-1 año)', 'Ambiente afectuoso genera confianza básica.'),
    ('E2a', 'Autonomía vs. Vergüenza\n(1-3 años)', 'Vestirse o recoger juguetes desarrolla autonomía.'),
    ('E3a', 'Iniciativa vs. Culpa\n(3-6 años)', 'Juegos creativos estimulan iniciativa.'),
    ('E4a', 'Laboriosidad vs. Inferioridad\n(6-12 años)', 'Reconocimiento escolar fomenta esfuerzo.'),
    ('E5a', 'Identidad vs. Confusión\n(12-18 años)', 'Proyectos vocacionales forman identidad.'),
    ('E6a', 'Intimidad vs. Aislamiento\n(18-40 años)', 'Trabajo colaborativo crea vínculos.'),
    ('E7a', 'Generatividad vs. Estancamiento\n(40-65 años)', 'Mentores transmiten saber.'),
    ('E8a', 'Integridad vs. Desesperación\n(65+ años)', 'Experiencias se comparten con jóvenes.'),
]

agrupaciones = {
    'cluster_infancia': ('Infancia y niñez', ['E1a', 'E2a', 'E3a', 'E4a']),
    'cluster_adolescencia': ('Adolescencia', ['E5a']),
    'cluster_adultez_temprana': ('Adultez temprana', ['E6a']),
    'cluster_adultez_media': ('Adultez media', ['E7a']),
    'cluster_adultez_tardia': ('Adultez tardía', ['E8a']),
}

# Añadir subgráficos para etapas
prev = None
for cluster_name, (label, ids) in agrupaciones.items():
    with dot.subgraph(name=cluster_name) as c:
        c.attr(label=label, style='dashed', color='blue')# style='filled', color='whitesmoke')
        for node_id in ids:
            etapa = next(e for e in etapas if e[0] == node_id)
            c.node(etapa[0], etapa[1], shape='box')
            ejemplo_id = 'ex_' + etapa[0]
            c.node(ejemplo_id, etapa[2], shape='note')
            c.edge(etapa[0], ejemplo_id)
            if prev:
                dot.edge(prev, etapa[0])
            else:
                dot.edge('D', etapa[0])
            prev = etapa[0]

# === Exportar y mostrar ===
output_path = dot.render('mapa_teoria_psicosocial_erikson_mejorado', cleanup=False)
Image(filename=output_path)
```

```python
from graphviz import Digraph
from IPython.display import Image, display  # Para notebooks

# Crear grafo principal
g = Digraph('DesarrolloHumano', filename='mapa_teorias_completo_vertical', format='png')
g.attr(rankdir='TB', size='20')

# Nodo central
g.node('DH', 'Desarrollo Humano\n(Proceso vital en etapas)', shape='ellipse', style='filled', color='lightgrey')

# Teorías principales
g.node('Freud', 'Teoría Psicosexual\n(Sigmund Freud)', shape='box', style='filled', color='lightcoral')
g.node('Erikson', 'Teoría Psicosocial\n(Erik Erikson)', shape='box', style='filled', color='lightblue')
g.node('Kohlberg', 'Teoría del Desarrollo Moral\n(Lawrence Kohlberg)', shape='box', style='filled', color='lightgreen')

# Conexión principal
g.edges([('DH', 'Freud'), ('DH', 'Erikson'), ('DH', 'Kohlberg')])

# === SUBGRAFO: Freud ===
with g.subgraph(name='cluster_Freud') as f:
    f.attr(label='Etapas Psicosexuales (Freud)', style='dashed', color='red')
    f.node('F1', 'Etapa Oral (0-1)\nPlacer en boca, alimentación')
    f.node('F2', 'Etapa Anal (1-3)\nControl de esfínteres')
    f.node('F3', 'Etapa Fálica (3-6)\nComplejo de Edipo')
    f.node('F4', 'Etapa de Latencia (6-12)\nRechazo de impulsos sexuales')
    f.node('F5', 'Etapa Genital (12+)\nDesarrollo sexual adulto')
    f.edges([('Freud', 'F1'), ('F1', 'F2'), ('F2', 'F3'), ('F3', 'F4'), ('F4', 'F5')])

# === SUBGRAFO: Erikson ===
with g.subgraph(name='cluster_Erikson') as e:
    e.attr(label='Etapas Psicosociales (Erikson)', style='dashed', color='blue')
    e.node('E1', 'Confianza vs. Desconfianza (0-1)')
    e.node('E2', 'Autonomía vs. Vergüenza (1-3)')
    e.node('E3', 'Iniciativa vs. Culpa (3-6)')
    e.node('E4', 'Laboriosidad vs. Inferioridad (6-12)')
    e.node('E5', 'Identidad vs. Confusión de roles (12-20)')
    e.node('E6', 'Intimidad vs. Aislamiento (20-40)')
    e.node('E7', 'Generatividad vs. Estancamiento (40-65)')
    e.node('E8', 'Integridad vs. Desesperanza (65+)')
    e.edges([('Erikson', 'E1'), ('E1', 'E2'), ('E2', 'E3'), ('E3', 'E4'),
             ('E4', 'E5'), ('E5', 'E6'), ('E6', 'E7'), ('E7', 'E8')])

# === SUBGRAFOS: Kohlberg ===
with g.subgraph(name='cluster_K1') as c:
    c.attr(label='Nivel Preconvencional (4-10 años)', style='dashed', color='green')
    c.node('K1', 'Nivel Preconvencional', shape='box', style='filled', color='lightgreen')
    c.node('K1a', '1. Obediencia y castigo')
    c.node('K1b', '2. Interés propio')
    c.edges([('K1', 'K1a'), ('K1', 'K1b')])

with g.subgraph(name='cluster_K2') as c:
    c.attr(label='Nivel Convencional (10-13 años)', style='dashed', color='green')
    c.node('K2', 'Nivel Convencional', shape='box', style='filled', color='lightgreen')
    c.node('K2a', '3. Concordancia interpersonal')
    c.node('K2b', '4. Autoridad y orden social')
    c.edges([('K2', 'K2a'), ('K2', 'K2b')])

with g.subgraph(name='cluster_K3') as c:
    c.attr(label='Nivel Postconvencional (13+)', style='dashed', color='green')
    c.node('K3', 'Nivel Postconvencional', shape='box', style='filled', color='lightgreen')
    c.node('K3a', '5. Contrato social')
    c.node('K3b', '6. Principios éticos universales')
    c.edges([('K3', 'K3a'), ('K3', 'K3b')])

# Conectar niveles a Kohlberg
g.edges([('Kohlberg', 'K1'), ('K1', 'K2'), ('K2', 'K3')])

# === Conexiones entre teorías ===
g.edge('F1', 'E1', style='dashed', label='Apego temprano')
g.edge('F2', 'E2', style='dashed', label='Autonomía corporal')
g.edge('F3', 'E3', style='dashed', label='Iniciativa ↔ deseo')
g.edge('F4', 'E4', style='dashed', label='Interés por logros')
g.edge('F5', 'E5', style='dashed', label='Identidad sexual')

g.edge('E4', 'K1', style='dashed', label='Edad escolar → moral')
g.edge('E5', 'K2', style='dashed', label='Adolescencia → moral social')
g.edge('E6', 'K3', style='dashed', label='Compromiso ético')
g.edge('E7', 'K3', style='dashed', label='Generatividad ética')
g.edge('E8', 'K3', style='dashed', label='Reflexión y justicia')

# Título general
g.attr(label='Mapa Conceptual: Freud – Erikson – Kohlberg\nInterconexión de Teorías del Desarrollo Humano', labelloc='t', fontsize='20')

# Renderizar
g.render(view=True)

# Mostrar si estás en Jupyter
try:
    display(Image('mapa_teorias_completo_vertical.png'))
except:
    pass
```

Para dibujar los mapas conceptuales, simplemente ejecuta los scripts de Python. Se generarán archivos de imagen en formato PNG con los mapas conceptuales.
